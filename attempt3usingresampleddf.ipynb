```python
import pandas as pa
import numpy as np

#according to Ted we should separate data into 60% and 40%, in order to train the 60% and test it against the 40%
before_sampling = pa.read_csv("/home/drey/Desktop/wells/merged.csv")

import random

#below was taken from http://stackoverflow.com/questions/12190874/pandas-sampling-a-dataframe, see first reply

rows = random.sample(before_sampling.index, int(59400.00*0.40))

after_sampling_40 = before_sampling.ix[rows]

after_sampling_60 = before_sampling.drop(rows)
```

```python
#the idea here is to assign a numeric value to functional, non functional, needs repair, and then find the average "status" for each source_type
after_sampling_60['source_type']
```

```python
#using the set functionality to get a unique list of the source types
sourcetype_list = list(set(after_sampling_60['source_type']))
```

```python
#creating a dictionary to assign a numeric value to source type - NOTE I don't actually use this below, yet
keyswithvalues = zip(sourcetype_list,range(0,len(sourcetype_list)))
sourcetype_dict = {key: value for (key,value) in keyswithvalues}
```

```python
#isolating out the only 2 columns I'm going to work on at the moment:
after_sampling_60_only2col = after_sampling_60[['source_type','status_group']]
```

```python
#(manually) created a dictionary to assign numeric values to the 3 status group
statusgroup_dict = {'functional':2 , 'functional needs repair':1, 'non functional':0}
```

```python
#creating a nested dictionary, as this is required as an argument for the .replace function in pandas.  
dict_reconfig = {'status_group':statusgroup_dict}
```

```python
#see note above - the replace function takes the first key in the dictionary as the column it should apply the value mappings (status group dict) to
A_S_60_numberfied = after_sampling_60_only2col.replace(to_replace=dict_reconfig)
```

```python
A_S_60_numberfied.dtypes
```

```python
#grouping by the source type and computing the average status group value associated with each
#NOTE the "as_index" parameter is required in order to output a usable NEW, GROUPED DF
A_S_60_groupedmeans = A_S_60_numberfied.groupby('source_type',as_index=False).aggregate(np.mean)
```

```python
#see? it worked!
A_S_60_groupedmeans
```

```python
#now I'm going to replace the source_type text in the DF with the status_group mean value from above table
meandictzipped = zip(A_S_60_groupedmeans['source_type'],A_S_60_groupedmeans['status_group'])
mean_dict = {'source_type':{key: value for (key,value) in meandictzipped}}
print mean_dict
```

```python
#using the replace method again
originaldata2col_withmeans = after_sampling_60_only2col.replace(to_replace=mean_dict)
```

```python
originaldata2col_withmeans
```

```python
#applying the same replacement to the 40% DF as we did above for the 60% DF
#Still not 100% clear why it's useful to separate DF into a 40 and 60%, sent question to Ted
#Note, sklean needs pandas dataframes as inputs to work, I think
A_S_40_withmeans = pa.DataFrame(after_sampling_40.replace(to_replace=mean_dict))
A_S_40_withmeansonly = pa.DataFrame(A_S_40_withmeans['source_type'])
```
